# **Business Understanding**

## Latar Belakang
Rumah sakit ingin mengurangi biaya jangka panjang dengan melakukan **deteksi dini diabetes** pada pasien.  
Saat ini, diagnosis hanya bisa dilakukan setelah tes darah lengkap, yang memakan waktu dan biaya.  

## Tujuan
- Membuat model prediksi sederhana berbasis data klinis dasar (misalnya umur, BMI, tekanan darah).  
- Memberikan skor risiko diabetes sehingga bisa memutuskan siapa yang perlu pemeriksaan lebih lanjut.
